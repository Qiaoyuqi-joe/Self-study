```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        from collections import defaultdict
        cnt = defaultdict(int)

        # 统计 s 的频次
        for ch in s:
            cnt[ch] += 1
        
        # 减去 t 的频次
        for ch in t:
            cnt[ch] -= 1
        
        # 所有字符频次必须为 0
        for v in cnt.values():
            if v != 0:
                return False
        
        return True
```
