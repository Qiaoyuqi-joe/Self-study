```python
class Solution:
    def lengthofLongestSubstring(self,s:str,k:int) -> int:
        from collections import defaultdict
        left = 0
        ans = 0
        distinct = 0
        cnt = defaultdict(int)
        for right,ch in enumerate(s):
            if cnt[ch]==0:
                distinct+=1
            cnt[ch]+=1

            while distinct>k:
                cnt[s[left]]-=1
                if cnt[s[left]] == 0:
                    distinct-=1
                left+=1
            ans = max(ans,right-left+1)
        return ans

```
