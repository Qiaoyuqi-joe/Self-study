```python
#å…¸å‹ä¾‹é¢˜ å­—ç¬¦ä¸²å…¨æ’åˆ—
# ğŸ¯ é¢˜ç›®ï¼šå­—ç¬¦ä¸²çš„å…¨æ’åˆ—ç»™å®šä¸€ä¸ªä¸å«é‡å¤å­—ç¬¦çš„å­—ç¬¦ä¸²ï¼ˆæˆ–æ•°ç»„ï¼‰ï¼Œ
# æ‰¾å‡ºå®ƒå¯ä»¥ç»„æˆçš„æ‰€æœ‰æ’åˆ—ç»„åˆã€‚
# è¾“å…¥ç¤ºä¾‹: s = "abc"è¾“å‡ºç¤ºä¾‹: ["abc", "acb", "bac", "bca", "cab", "cba"]
# è§£å†³æ€è·¯ï¼šæˆ‘ä»¬ä½¿ç”¨å›æº¯æ³•ï¼Œç›®æ ‡æ˜¯æ„å»ºä¸€ä¸ªé•¿åº¦ä¸º $N$ï¼ˆå­—ç¬¦ä¸²é•¿åº¦ï¼‰çš„è·¯å¾„ã€‚
# åœ¨æ¯ä¸€æ­¥ï¼Œæˆ‘ä»¬éƒ½è¦ä»åŸå­—ç¬¦ä¸²ä¸­é€‰æ‹©ä¸€ä¸ªå°šæœªä½¿ç”¨çš„å­—ç¬¦æ·»åŠ åˆ°å½“å‰è·¯å¾„ä¸­ã€‚
```

```python
def solve(self,chars):
    self.results = []#æœ€ç»ˆæ‰¾åˆ°çš„æ‰€æœ‰æ’åˆ—ç»“æœ

    self.path = []#å½“å‰æ­£åœ¨æ„å»ºçš„æ’åˆ— è¿™æ˜¯ä¸€ä¸ªæ ¸å¿ƒå˜é‡ï¼Œå®ƒä»£è¡¨äº†æˆ‘ä»¬ä»èµ·å§‹ç‚¹èµ°åˆ°å½“å‰æ·±åº¦æ‰€å½¢æˆçš„æ’åˆ—ç‰‡æ®µã€‚åœ¨é€’å½’è¿‡ç¨‹ä¸­ä¸æ–­ appendï¼Œå›æº¯æ—¶ pop
    self.used = [False] * len(chars)#æ ‡è®°æ•°ç»„ ç”¨æ¥è®°å½•å“ªä¸€ä¸ªè¢«ä½¿ç”¨äº†
    self._backtrack(chars)
    return self.results

def _backtrack(self,chars):
    #ç»ˆæ­¢æ¡ä»¶ å½“è·¯å¾„çš„é•¿åº¦ç­‰äºå­—ç¬¦é•¿åº¦æ€»æ•° æ‰¾åˆ°ä¸€ä¸ªå®Œæ•´çš„æ’åˆ—
    if len(self.path) == len(chars):
        self.results.append("".join(self.path))#è½¬æ¢æˆå­—ç¬¦ä¸²å½¢å¼ åŠ å…¥åˆ—è¡¨ä¹‹ä¸­
        return #æ ‡å¿—è¿™æ¡è·¯å¾„ç»“æŸ å°†é€€å›åˆ°ä¸Šä¸€ä¸ªé€’å½’è°ƒç”¨å±‚ å¼€å§‹æ‰§è¡Œè¯¥å±‚ä¸­çš„æ’¤é”€é€‰æ‹©
        #å½“returnå®ç°ä¹‹å æ‰ä¼šè¿›è¡Œpopï¼ˆï¼‰ å› ä¸ºreturn ç»“æŸçš„æ˜¯å½“å‰è¿™ä¸ªå‡½æ•°çš„æ‰§è¡Œ æŠŠæ§åˆ¶æƒäº¤ç»™è°ƒç”¨çš„ä¸Šä¸€å±‚å‡½æ•° åˆ™å°±ç»§ç»­æ‰§è¡Œä¸Šä¸€å±‚çš„pop
    #éå†æ‰€æœ‰çš„å¯èƒ½çš„é€‰æ‹©
    for i in range(len(chars)):
        #å¦‚æœchars[i]å·²ç»è¢«ä½¿ç”¨ åˆ™è·³è¿‡
        if self.used[i]:
            continue
        #åšå‡ºé€‰æ‹©
        char = chars[i]
        self.path.append(char)
        #æ ‡è®°è¯¥å­—ç¬¦ä¸ºå·²ä½¿ç”¨
        self.used[i] = True

        #é€’å½’è°ƒç”¨ ç»§ç»­æ¢ç´¢ä¸‹ä¸€ä¸ªä½ç½®å¯èƒ½æ€§å’Œä¸‹ä¸€ä¸ªå­—ç¬¦çš„é€‰æ‹© å°±è¿›å…¥ä¸‹ä¸€å±‚äº†
        self._backtrack(chars)

        #æ’¤é”€é€‰æ‹©  é€’å½’è¢«è°ƒç”¨çš„æ—¶å€™ å°±æ„å‘³ç€è¯¥åˆ†æ”¯æœç´¢å®Œæ¯• æ‰€ä»¥è¦æ’¤é”€ æ¢å¤åˆ°é€’å½’ä»¥å‰ 
        self.path.pop()
        self.used[i] = False
```
