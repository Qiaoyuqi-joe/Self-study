单调栈是一种特殊的栈数据结构，其核心特性是栈内元素始终保持单调递增或单调递减。通过维持这种单调性，它能高效解决「找到下一个更大 / 更小元素」类问题，时间复杂度通常为 O(n)（每个元素仅入栈和出栈一次）。




单调栈本质 是记录已经读过的数据



```python
def dailyTemperatures(temperatures):
    n = len(temperatures)
    answer = [0]*n
    stack = []#初始化栈 存储下标
    for i in range(n):
        while stack and temperatures[i]>temperatures[stack[-1]]:
            top = stack.pop()
            answer[top] = i-top
        stack.append(i)
    return answer

```

```python
from typing import List

class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        n = len(temperatures)
        result = [0] * n
        stack = []  # 存储下标，栈内温度单调递减
        
        for i in range(n):
            # 当栈不为空，且当前温度 > 栈顶下标对应的温度时，弹出栈顶并计算天数
            while stack and temperatures[i] > temperatures[stack[-1]]:
                top = stack.pop()  # 修正：使用 pop() 方法弹出栈顶元素
                result[top] = i - top
            stack.append(i)  # 当前下标入栈
        return result
```

stack[-1]获取栈顶元素 但不删除
stack.pop() 移除并返回栈顶元素
